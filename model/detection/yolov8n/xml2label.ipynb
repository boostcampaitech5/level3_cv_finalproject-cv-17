```python
import os
import shutil
import xml.etree.ElementTree as ET
```

```python
root = './val_labeling'
img_root = './data/val/images'
save = './data/val/labels'

img_set = set(os.listdir(img_root))

for filename in os.listdir(root):
    xml_path = os.path.join(root, filename)
    tree = ET.parse(xml_path)
    root_element = tree.getroot()
    image_elements = root_element.findall('image')
    
    for image_element in image_elements:
        # <image> 엘리먼트의 name 속성 추출
        image_name = image_element.get('name')
        if image_name not in img_set:
            continue
        
        image_width = float(image_element.get('width'))
        image_height = float(image_element.get('height'))
        print(f"Image name: {image_name}")

        left_eyelid = [0]
        right_eyelid = [1]
        # <points> 엘리먼트 추출
        polygons_elements = image_element.findall('polygon')
        for polygons_element in polygons_elements:
            # <points> 엘리먼트의 points 속성 추출
            if polygons_element.get('label') == "left_eyelid":
                X = []
                Y = []
                points = polygons_element.get('points')
                points = points.split(';')
                for point in points:
                    x,y = point.split(',')
                    X.append(int(round(float(x))))
                    Y.append(int(round(float(y))))
                x1 = min(X)
                y1 = min(Y)
                x2 = max(X)
                y2 = max(Y)
                xm = (x1+x2)/2 /image_width 
                ym = (y1+y2)/2 /image_height
                xg = (x2-x1) /image_width  *2
                yg = (y2-y1) /image_height *4
                left_eyelid.append(round(xm, 6))
                left_eyelid.append(round(ym, 6))
                left_eyelid.append(round(xg, 6))
                left_eyelid.append(round(yg, 6))
                
            if polygons_element.get('label') == "right_eyelid":
                X = []
                Y = []
                points = polygons_element.get('points')
                points = points.split(';')
                for point in points:
                    x,y = point.split(',')
                    X.append(int(round(float(x))))
                    Y.append(int(round(float(y))))
                x1 = min(X)
                y1 = min(Y)
                x2 = max(X)
                y2 = max(Y)
                xm = (x1+x2)/2 /image_width 
                ym = (y1+y2)/2 /image_height
                xg = (x2-x1) /image_width  *2
                yg = (y2-y1) /image_height *4
                right_eyelid.append(round(xm, 6))
                right_eyelid.append(round(ym, 6))
                right_eyelid.append(round(xg, 6))
                right_eyelid.append(round(yg, 6))
                
        print("left_eyelid",left_eyelid)
        print("right_eyelid",right_eyelid)
        
        str_left_eyelid = " ".join(map(str,left_eyelid))
        str_right_eyelid= " ".join(map(str,right_eyelid))
        
        image_name = image_name[:-3] + 'txt'
        save_path = os.path.join(save, image_name)
        with open(save_path, "w") as file:
            file.write(str_left_eyelid)
            file.write("\n")
            file.write(str_right_eyelid)
```
